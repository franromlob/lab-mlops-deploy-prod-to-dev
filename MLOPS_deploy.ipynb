```python
# --- Imports ---
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- Simple mathematical operation ---
# Create an array of numbers from 0 to 9
x = np.arange(0, 10)

# Compute a simple mathematical function
y = x ** 2  # square of each number

# Convert to DataFrame (just to use pandas)
df = pd.DataFrame({"x": x, "y": y})
print("DataFrame with values:")
print(df)

# --- Plot ---
plt.figure(figsize=(6, 4))
plt.plot(x, y, marker='o', color='blue')
plt.title("Function y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
